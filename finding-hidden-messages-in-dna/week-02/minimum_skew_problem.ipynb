```
import requests
def get_organism_genome(url):   
  genome_e_coli_enome = ""
  try:
      r = requests.get(url)
      genome_e_coli_enome = r.text
      return genome_e_coli_enome
  except requests.exceptions.HTTPError as e:
      print (e.response.text)
      return ""
```

```
def minimum_skew_problem(genome):
  c = 0
  g = 0
  minskew = 0
  skew_list = []
  index = 0
  for base in genome:
    index += 1
    if base == "C":
      c+=1				
    if base == "G":
      g+=1
    skewindex = g-c
    if skewindex < minskew:  
      skew_list = [index]
      minskew = skewindex
    if skewindex == minskew and index not in skew_list:
      skew_list.append(index)
  print(skew_list)


genome = get_organism_genome("https://raw.githubusercontent.com/ovvesley/bioinformatics/master/genomes/skewproblemminnum.txt")
minimum_skew_problem(genome)

  
```

```
def skew(sequence):
    c = 0
    g = 0
    min_skew = 0
    skew_list = []
    index = 0
    for i in sequence:
        index += 1
        if i == 'C':
            c += 1
        if i == 'G':
            g += 1
        skew = g-c
        if skew < min_skew:
            skew_list = [index]
            min_skew = skew
        if skew == min_skew and index not in skew_list:
            skew_list.append(index)    
    print(skew_list)

genome = get_organism_genome("https://raw.githubusercontent.com/ovvesley/bioinformatics/master/genomes/extra_skew.txt")
skew(genome)
```
